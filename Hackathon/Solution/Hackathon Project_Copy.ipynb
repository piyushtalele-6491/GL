```python
# Creating a new dataframe without the principal component dimensions
df_nPC = df[['MEAN_RR', 'MEDIAN_RR', 'SDRR', 'RMSSD', 'SDSD', 'SDRR_RMSSD',
       'pNN25', 'pNN50', 'KURT', 'SKEW', 'MEAN_REL_RR', 'MEDIAN_REL_RR',
       'SDRR_REL_RR', 'RMSSD_REL_RR', 'SDSD_REL_RR', 'SDRR_RMSSD_REL_RR',
       'KURT_REL_RR', 'SKEW_REL_RR', 'VLF', 'LF', 'LF_NU',
       'HF', 'HF_NU', 'TP', 'LF_HF', 'HF_LF', 'SD1', 'SD2', 'sampen',
       'higuci', 'no stress', 'interruption',
       'time pressure', 'HR']]
```

```python
df_nPC
```

```python
X = df_nPC.iloc[:,:-1]
```

```python
y = pd.DataFrame(df_nPC['HR'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data into Train/Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=76)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Applying Standard Scaler to the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# Applying PCA
from sklearn.decomposition import PCA
# Limiting the number of components to 10
pca = PCA(n_components=10)
X_train_PCA = pca.fit_transform(X_train_scaled)
X_test_PCA = pca.transform(X_test_scaled)
print(pca.explained_variance_)
print(pca.explained_variance_ratio_)
```

#### Using Random forest

```python
%%time
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
regr = RandomForestRegressor()
print(datetime.now().strftime("%d/%m/%Y %H:%M:%S"))
regr.fit(X_train_PCA, y_train.values.ravel())
y_pred_rf = regr.predict(X_test_PCA)
print('Random Forest Performance Parameters')
print('Training Score: {}'.format(regr.score(X_train_PCA, y_train)))
print('Testing Score: {}'.format(regr.score(X_test_PCA, y_test)))
print('MSE: {}'.format(mean_squared_error(y_test, y_pred_rf)))
print('MAE: {}'.format(mean_absolute_error(y_test, y_pred_rf)))
print(datetime.now().strftime("%d/%m/%Y %H:%M:%S"))
```

```python
regr.get_params(deep=True)
```

#### Using ADABoost Regressor

```python
from sklearn.ensemble import AdaBoostRegressor
adaregr = AdaBoostRegressor(n_estimators=100)
print(datetime.now().strftime("%d/%m/%Y %H:%M:%S"))
adaregr.fit(X_train_PCA, y_train.values.ravel())
y_pred_adab = adaregr.predict(X_test_PCA)
print('ADABoost Performance Parameters')
print('Training Score: {}'.format(adaregr.score(X_train_PCA, y_train)))
print('Testing Score: {}'.format(adaregr.score(X_test_PCA, y_test)))
print('MSE: {}'.format(mean_squared_error(y_test, y_pred_adab)))
print('MAE: {}'.format(mean_absolute_error(y_test, y_pred_adab)))
print(datetime.now().strftime("%d/%m/%Y %H:%M:%S"))
```

```python
adaregr.get_params(deep=True)
```

#### Using Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
gbregr = GradientBoostingRegressor(min_samples_leaf=5)
print(datetime.now().strftime("%d/%m/%Y %H:%M:%S"))
gbregr.fit(X_train_PCA, y_train.values.ravel())
y_pred_gbr = gbregr.predict(X_test_PCA)
print('Gradient Boost Performance Parameters')
print('Training Score: {}'.format(gbregr.score(X_train_PCA, y_train)))
print('Testing Score: {}'.format(gbregr.score(X_test_PCA, y_test)))
print('MSE: {}'.format(mean_squared_error(y_test, y_pred_gbr)))
print('MAE: {}'.format(mean_absolute_error(y_test, y_pred_gbr)))
print(datetime.now().strftime("%d/%m/%Y %H:%M:%S"))
```
